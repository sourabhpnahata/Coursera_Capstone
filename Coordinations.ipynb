<h4>Accessing the files</h4>

```python
df = pd.read_csv('Toronto_df.csv')
df1 = pd.read_csv('Geospatial_Coordinates.csv')
```

```python
df['Postcode'].unique()
```

<h4> Sorting according to Postcode</h4>

```python
df = df.sort_values(by = ['Postcode'], ascending = True)
df.reset_index(inplace= True, drop = True)
df.head()
```

<h4> Merging data from both Data frames</h4>

```python
df =pd.merge(df, df1, how='left', left_on='Postcode', right_on='Postal Code',
         sort=True,suffixes=('_x', '_y'), copy=True, indicator=False,
         validate=None)
```

<h4> Selecting Columns which we are intrested in </h4>

```python
df = df[['Postcode','Borough','Neighbourhood','Latitude','Longitude']]
```
